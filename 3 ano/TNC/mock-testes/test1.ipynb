Sejam n=2^{30}-1 e m=2^{29}-1 , e considere o isomorfismo de anéis f de Znm para ZnxZm, tal como definido nas aulas. Calcule a imagem recíproca, por f, da soma de (a_1, a_2) com (b_1, b_2), tomando (a_1, a_2)= (491461630, 377404666) e (b_1, b_2)= (853868122, 364720484).

```python
n = 2^30 - 1
m = 2^29 - 1
a_1, a_2 = 491461630, 377404666
b_1, b_2 = 853868122, 364720484

Zn = IntegerModRing(n)
Zm = IntegerModRing(m)
Znm = IntegerModRing(n*m)

s1 = Zn(a_1 + b_1)        # resto em Z_n
s2 = Zm(a_2 + b_2)        # resto em Z_m

x  = crt(ZZ(s1), ZZ(s2), n, m)   # converter para inteiros!
x_mod = Znm(x)
x_mod
```
