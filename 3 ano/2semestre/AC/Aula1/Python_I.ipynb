### Text type (strings)

Text type is also one of the most basic and important data types in Python.

It is used to represent text data. It represents a sequence of characters.

Strings in python are surrounded by either single quotation marks, or double quotation marks.

'hello' is the same as "hello".

You can assign a multiline string to a variable by using three quotes:

```python
print("Hello")
print('Hello')

a = "Hello"
print(a)

a = """This is a
multiline
string!"""
print(a)
```

Python has a set of built-in methods that you can use on strings.

Check [https://www.w3schools.com/python/python_ref_string.asp](https://www.w3schools.com/python/python_ref_string.asp) if you want to know more.

Some examples:

```python
dna = "  ATCGgaaa  "

# Cleanup & casing
print(dna.strip())     # remove surrounding spaces
print(dna.upper())     # uppercase
print(dna.lower())     # lowercase

print(len(dna))        # total number of characters (including spaces)
```

```python
print("aaabbc".count('a')) # count number of "a" in the string
```

```python
# Replace & split
print(dna.replace("A", "X"))
seq = "GGA-CTT-AAA"
print(seq.split("-"))
```

```python
# f-strings (formatted strings)
name, mw = "Glucose", 180.16

print(f"Default: The molecule {name} has molecular weight {mw} g/mol")

# Floating-point formats
print(f"Fixed point (2 decimals): {mw:.2f} g/mol")
print(f"Fixed point (0 decimals): {mw:.0f} g/mol")
print(f"Scientific notation (2 decimals): {mw:.2e} g/mol")

# Adding commas
print(f"With comma as thousands separator: {1234567.89:,.2f}")

# Percentages
percent = 0.12345
print(f"Percentage (default): {percent:%}")
print(f"Percentage (1 decimal): {percent:.1%}")
```

```python
# Escape sequences and raw strings
print("Line1 \nLine2")     # newline
print(r"Raw string keeps backslashes like \nLine2")
```

```python
# What NOT to do: comparing strings with 'is' instead of '=='
s1 = "ATP"
s2 = "".join(["A","T","P"])
print("Equality (==):", s1 == s2)
print("Identity  (is):", s1 is s2)
```

# Python Conditions and If statements

In Python, if statements are used to perform different actions based on different conditions.

Python supports the usual logical conditions from mathematics:

- Equal: `a == b`
- Not Equal: `a != b`
- Less than: `a < b`
- Less than or equal to: `a <= b`
- Greater than: `a > b`
- Greater than or equal to: `a >= b`

These conditions can be used in several ways, most commonly in "if statements" and loops.

An "if statement" is written by using the `if` keyword.

```python
a = 33
b = 200
if b > a:
  print("b is greater than a")
if a > b:
  print("a is greater than b")
```

### Indentation

Python relies on indentation (whitespace at the beginning of a line) to define scope in the code.

If statements must be indented to tell Python which part of the code is part of the statement.


```python
a = 33
b = 200
if b > a:
print("b is greater than a") # you will get an error if you dont use the correct identation
```

### Elif

The `elif` keyword is pythons way of saying "if the previous conditions were not true, then try this condition".

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

### Else

The `else` keyword catches anything which isn't caught by the preceding conditions.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

### Logical Operators

Logical operators can be used to combine conditions. They are `and`, `or`, and `not`.


#### And

The `and` keyword is a logical operator, and is used to combine conditional statements:

```python
# Test if a is greater than b, AND if c is greater than a:

a = 200
b = 33
c = 150

if a > b and c > a:
  print("Both conditions are True")
```

### Or

The `or` keyword is a logical operator, and is used to combine conditional statements:

```python
# Test if a is greater than b, OR if a is greater than c:

a = 200
b = 33
c = 500
if a > b or a > c or b > c:
  print("At least one of the conditions is True")
```

### Not

The `not` is used to invert the logic of a conditional statement:

```python
# Test if b is NOT greater than a using the `not` keyword:

a = 200
b = 33
c = 500
if not b > a: # b > a is False but with the `not` operator it is inverted to True
  print("B is not greater than A")
```

# Python Loops

Python has two primitive loop commands:

- `while` loops
- `for` loops

In Python, loops are used to iterate over collections of data or to repeat a statement a given number of times.

### The while Loop

With the `while` loop we can execute a set of statements as long as a condition is true.

The `while` loop is very useful for executing a set of statements for an unknown number of times.

```python
i = 1
while i < 6:
  print("i number: ", i)
  i += 1 # i += 1 is the same as i = i + 1

print("i number at the end:", i)
```

#### The `break` Statement

With the `break` statement we can stop the loop even if the while condition is true:

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

#### The `continue` Statement

With the `continue` statement we can stop the current iteration, and continue with the next:

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

#### The `else` Statement in a `while` loop

With the `else` statement we can run a block of code once when the condition no longer is true:

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

### The For Loop


The `for` loop is used to iterate over a sequence

This is less like the `for` keyword in other programming languages, and works more like an iterator method

In most languages, for means “count from A to B” — in Python, for means “iterate over things.”

#### The `range()` Function

To loop through a set of code a specified number of times, we can use the `range()` function,

The `range()` function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```python
for i in range(10):
  print(i)
```

```python
for i in range(1, 10):
  print(i)
```

```python
# break & continue
for i in range(1, 10):
    if i == 5:
        break      # stop loop entirely
    if i % 2 == 0:
        continue   # skip even numbers
    print("odd:", i)
```

```python
for x in "banana":
  print(x)
```

The `range()` function defaults to increment the sequence by 1, however it is possible to specify the increment value by adding a third parameter: `range(2, 30, 3)`.

```python
for x in range(2, 30, 3):
  print(x)
```

```python
for i in range(2, 1):     #what does this print? and what if we add: ,-1)
    print(i)
```
