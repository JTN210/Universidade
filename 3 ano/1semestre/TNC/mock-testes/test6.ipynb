Sejam n=2^{30}-1 e m=2^{29}-1 , e considere o isomorfismo de anéis f de Znm para ZnxZm, tal como definido nas aulas. Calcule a imagem recíproca, por f, da soma de (a_1, a_2) com (b_1, b_2), tomando (a_1, a_2)= (273933845, 472000352) e (b_1, b_2)= (976055371, 296993273) .

```sage
n = 2^30 - 1  
m = 2^29 - 1 
a_1, a_2 = 273933845, 472000352
b_1, b_2 = 976055371, 296993273

Zn = IntegerModRing(n)
Zm = IntegerModRing(m)
Znm = IntegerModRing(n*m)

s1 = Zn(a_1 + b_1)
s2 = Zm(a_2 + b_2)

x = crt(ZZ(s1),ZZ(s2),n,m)
x_mod = Znm(x)
x_mod
```
